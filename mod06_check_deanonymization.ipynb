# Discussion Questions

### Describe your reaction to the match rate. What sources of data might be prone to privacy issues?

The match rate shows that even when direct identifiers are removed, people can still sometimes be re-identified using a small set of quasi-identifiers like age, gender, and ZIP code. This is concerning because it means anonymized data may still pose privacy risks when combined with outside datasets.Sources that contain demographic or location-related information are especially vulnerable, such as voter registration records, public records, social media profiles, health datasets, consumer databases, and location/app data. When these sources share overlapping attributes, linking becomes easier.
